```python
d = {}
for i in data :

    if str(i[3]).strip() == 'Unclaimed' : continue 

    key = i[6] # directory 
    if str(key).count('ICF/automated/') > 0 :
        key = key.split('ICF/automated/')[1]

    # ntuple containts: job id, user, release, tag set, status, dataset name
    if key in d.keys() :
        d[key].append((i[0],i[5],i[1],i[2],i[4],i[7]))
    else :
        d[key] = [(i[0],i[5],i[1],i[2],i[4],i[7])]
        
```
